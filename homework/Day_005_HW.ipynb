```python
#平均數 mean
print('AMT_INCOME_TOTAL 平均數:'+ str(app_train['AMT_INCOME_TOTAL'].mean()))
#標準差 std
print('AMT_INCOME_TOTAL 標準差:'+ str(app_train['AMT_INCOME_TOTAL'].std()))
```

```python
# 計算有沒有車
ownCar  =  (app_train['FLAG_OWN_CAR'] == 'Y').sum()
noCar  =  (app_train['FLAG_OWN_CAR'] == 'N').sum()
print(ownCar)
```

```python
x = ['YES', 'NO']
carNum = [ownCar, noCar]
plt.bar(x, carNum, label = 'carNum')
```
